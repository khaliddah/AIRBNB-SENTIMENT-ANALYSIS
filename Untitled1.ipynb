```python
#on se base sur l'anglais
reviews = reviews[(reviews['language']=='en')]
```

```python
from nltk.corpus import stopwords
reviews = reviews[reviews['comments'].notnull()]
stopwords_list = set(stopwords.words("english"))
punctuations = """!()-![]{};:,+'"\,<>./?@#$%^&*_~Â""" #List of punctuation to remove

def reviewParse(review):
    splitReview = review.split() #Split the review into words
    parsedReview = " ".join([word.translate(str.maketrans('', '', punctuations)) + " " for word in splitReview])
    #Takes the stubborn punctuation out
    return parsedReview #Returns the parsed review

def clean_review(review):
    clean_words = []
    splitReview = review.split()
    for w in splitReview:
        if w.isalpha() and w not in stopwords_list:
            clean_words.append(w.lower())
    clean_review = " ".join(clean_words)
    return clean_review
```

```python
#appliquer les 2 methodes sur la colonne comments
reviews["comments"] = reviews["comments"].apply(reviewParse).apply(clean_review)
```

# Classification bayesienne

```python
from nltk.tokenize import word_tokenize 
#word_tokenize est une fonction en Python qui divise une phrase donnée en mots en utilisant la bibliothèque NLTK.
df["comments"] = df["comments"].apply(lambda x: word_tokenize(x))
```

```python
df["comments"].iloc[0]
```

```python
# iterrows() est utilisé pour faire des iterations sur les lignes 
# d'un cadre de données pandas sous la forme d'une paire.
documents = []
for index,row in df.iterrows():
    documents.append((row['comments'], row['sentiment']))
```

```python
import random
random.shuffle(documents)

# affichage du premier element de la liste (review + categorie) 
print(documents[0])
```

```python
import nltk
#variable contient la fréquence d'apparition de chaque mot 
all_words = nltk.FreqDist(s.lower() for w in df["comments"] for s in w)
#dans un premier lieux on limite le nombre de caracteristique à 1000 mots les plus frequents
word_features = list(all_words)[:1000]
print(word_features)
```

```python
from nltk.corpus import stopwords
import string
def clean_doc(doc):
    # supprimer la ponctuation de chaque token
    table = str.maketrans('', '', string.punctuation)
    tokens = [w.translate(table) for w in doc]
    # supprimer les tokens restants qui ne sont pas alphabétiques
    tokens = [word for word in tokens if word.isalpha()]
    #filtrer les mots vides
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if not w in stop_words]
    # filtrer les tokens courts
    tokens = [word for word in tokens if len(word) > 1]
    return tokens
 
    
# charger le document
word_features = clean_doc(word_features)
print(word_features)
```

```python
## une fonction qui vas retourner un dictionnaire et qui teste l'existance
# de chaque mots dans word_features par rapport a l'ensemble du document.
def document_features(document):
    #on a utilise set(document) pour que les mots du document
    #ne soient pas repetetifs
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words)
    return features
```

```python
# Train Naive Bayes classifier pour prédire les sentiments(catégories) des nouveaux movie reviews.
featuresets = [(document_features(d), c) for (d,c) in documents]
print(featuresets[0])
train_set, test_set = featuresets[100:], featuresets[:100]
print(np.shape(train_set))#95%
print(np.shape(test_set))#5%
```

```python
# Train Naive Bayes
classifier = nltk.NaiveBayesClassifier.train(train_set)
# Test the classifier
print(nltk.classify.accuracy(classifier, test_set))
```

```python
# # Montrer les caractéristiques les plus importantes 
# telles qu'interprétées par Naive Bayes.
classifier.show_most_informative_features(5)
```
